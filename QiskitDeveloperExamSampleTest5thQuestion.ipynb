# Bell State: Maximally entangled state 


```python
#Import the required packages
from qiskit import QuantumCircuit, Aer,execute
from qiskit.visualization import *
```

```python
#Create a Quantum Circuit with 2 qubits and 2 classical bits
qc_bell=QuantumCircuit(2,2)
qc_bell.h(0)
qc_bell.cx(0,1)

#Measurement
qc_bell.measure([0,1],[0,1])
qc_bell.draw('mpl')
```

```python
#Get the qasm simulator from Aer
backend=Aer.get_backend('qasm_simulator')

#Execute the circuit
#shots= number of times the experiment/circuit is run

job=execute(qc_bell,backend,shots=1024)

#Get the job result: result()

result=job.result()

#Get the histogram data of the experiement

counts=result.get_counts()

print(counts)
```

## Plot the histogram data
- Output obtained is the probabities.
- Probabilities are calculated by taking the individual counts of the state / shots

```python
plot_histogram(counts,title="Bell State without noise")
```

## Conclusion:

-The above qasm simulator doesn't contain any noise, So output obtained is without noise
- This is called as maximally entangled quantum state of two qubits

The Quantum circuit created is called a Bell state.
- If Both qubits are 50 % having the value |00>
- If Both qubits are 50 % having teh value |11>

# 4 types of maximally entangled states: 2-qubit bell state


INPUT 00:-----> $|\Phi^+\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle + |11\rangle\right)$

INPUT 01:----->$|\Phi^-\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle - |11\rangle\right)$

INPUT 10:----->$\vert\Psi^+\rangle = \frac{1}{\sqrt{2}}\left(\vert01\rangle + \vert10\rangle\right)$

INPUT 11:----->$\vert\Psi^-\rangle = \frac{1}{\sqrt{2}}\left(\vert01\rangle - \vert10\rangle\right)$
 

# Example 1: input=00

$|\Phi^+\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle + |11\rangle\right)$

```python
qc1_bell=QuantumCircuit(2)
qc1_bell.h(0)
qc1_bell.cx(0,1)
qc1_bell.draw(output="mpl")
```

```python
backend=Aer.get_backend('statevector_simulator')
job=execute(qc1_bell,backend,shots=1024)
result=job.result()
sv=result.get_statevector()
sv.draw(output="latex")
```

# Example 2: input=01

$|\Phi^-\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle - |11\rangle\right)$

```python
qc2_bell=QuantumCircuit(2)
qc2_bell.x(0)
qc2_bell.h(0)
qc2_bell.cx(0,1)

qc2_bell.draw(output="mpl")
```

```python
backend=Aer.get_backend('statevector_simulator')
job=execute(qc2_bell,backend,shots=1024)
result=job.result()
sv=result.get_statevector()
sv.draw(output="latex")
```

### NOTE: Example 2 is same as Example 1 but with a negative phase

### Let's see qsphere representation of a quantum state.Color represents the phase

```python
plot_state_qsphere(sv,show_state_phases=True)
```
